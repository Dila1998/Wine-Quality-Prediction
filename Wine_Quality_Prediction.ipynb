import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

data collection

```python
#loading the dataset
wine_dataset = pd.read_csv('/content/winequality-red.csv')
```

```python
# the 1st 5rows of data set
wine_dataset.head()
```

```python
#checking missing values
wine_dataset.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'wine_dataset' is a pandas DataFrame

for column in wine_dataset.columns:
    if column != 'quality':  # Skip the 'quality' column itself
        plt.figure(figsize=(5, 5))  # Create a new figure for each plot
        sns.barplot(x='quality', y=column, data=wine_dataset)
        plt.title(f'{column} vs. Wine Quality')  # Set title with column name
        plt.xlabel('Wine Quality')
        plt.ylabel(column)
        plt.show()
```

data pre processing

```python
#separate data and lable
X = wine_dataset.drop('quality', axis=1)
print(X)
```

lable binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print (Y)
```

train and test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print (Y.shape, Y_train.shape, Y_test.shape)
```

model training:

random forest classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

model evaluation

accuracy score

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print ('Accuracy:', test_data_accuracy)
```

Bulding a predictive system

```python
input_data = (7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)

#changing the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
